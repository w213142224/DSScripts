**Q02. Given the following data, create a line chart with the following properties:**<br>
      
      1. Chart title:'Weekly Temperature'; label for X axis:'Week Day'; label for Y axis:'Tempearue in Fahrenheit'
      2. Add line color, line width, line style, marker, and alpha of your choice
      3. Add legend and grid
      4. Use output image as reference
**Output image:** <br>      


```python
week_day = ['SUN','MON','TUE','WED','THU','FRI','SAT']
high_temp = [88, 97, 95, 98, 100, 101, 93]
low_temp = [77, 80, 79, 73, 80, 81, 78]
avg_temp = [82, 88, 87, 85, 90, 91, 86]
plt.plot(week_day, high_temp, label='High', color='blue', linewidth=1, linestyle='--', marker='d', alpha=0.7)
plt.plot(week_day, low_temp, label='Low', color='green', linewidth=1, linestyle='-.', marker='+', alpha=0.7)
plt.plot(week_day, avg_temp, label='Avg', color='red', linewidth=1, linestyle='-', marker='.', alpha=0.7)

# Title
plt.title('Weekly Temperature')
plt.xlabel('Day of the Week')
plt.ylabel('Temperature in Fahrenheit')


plt.legend(loc='lower right')
plt.grid(True)

plt.show()
```

**Q03. Use the data below, create a stacked bar chart. Use the output image as reference**<br>
years = range(2015,2020) <br>
class_counts = {
    'Face to Face':np.array([47, 50, 48, 40, 35]),
    'Online': np.array([29, 28, 40, 45, 50]),
}<br><br>
**Output image:** <br>


```python
years = range(2015,2020)
class_counts = { 'Face to Face':np.array([47, 50, 48, 40, 35]), 'Online': np.array([29, 28, 40, 45, 50]), }
f2f = class_counts['Face to Face']
online = class_counts['Online']

# Plotting the bars
plt.bar(years, f2f, label='Face to Face')
plt.bar(years, online, bottom=f2f, label='Online')

# Adding text labels inside the bars to match the reference image
for i in range(len(years)):
    plt.text(years[i], f2f[i]/2, str(f2f[i]), ha='center', va='center')
    plt.text(years[i], f2f[i] + (online[i]/2), str(online[i]), ha='center', va='center')

# Labels and Title
plt.title('Class Modality')
plt.xlabel('Years')
plt.ylabel('Number of Classes')

# legend and grid
plt.legend(loc='upper left')
plt.grid(axis='y', linestyle='-', alpha=0.5)

plt.show()
```

**Q04. Load dataset 'iris' from Seaborn and create a stacked Histogram chart to see the distribution of sepal length. Use the output image as reference.** <br><br>
**Output image:**<br>


```python
import matplotlib.pyplot as plt
import seaborn as sns

flowers_df = sns.load_dataset("iris")

sns.set_style("darkgrid")

flowers_df['species'] = flowers_df['species'].str.capitalize()

sns.histplot(data=flowers_df, x='sepal_length', hue='species', multiple='stack')

plt.title('Distribution of Sepal Length')

plt.show()
```

**Q05. Load dataset 'tips' from Seaborn and create the bar chart (use Seaborn barplot() method) to display bill amount on each day based on gender of servers. Use the output image as reference.**<br><br>
**Output image:** <br>


```python
tips_df = sns.load_dataset('tips')
print(tips_df)

tips_df = sns.load_dataset('tips')

sns.set_style("whitegrid")

sns.barplot(data=tips_df, x='day', y='total_bill', hue='sex', palette=['blue', 'red'])

plt.title('Daily Bill Amount by Gender')
plt.xlabel('Day')
plt.ylabel('Total Bill')

plt.legend(title='sex', loc='upper left')

plt.show()
```

**Q06. Convert the Bar chart from Q05 to Horizontal Bar chart. Use the output image as reference.**<br><br>
**Output image:** <br>


```python
sns.set_style("whitegrid")

sns.barplot(data=tips_df, x='total_bill', y='day', hue='sex', palette=['blue', 'red'])

plt.title('Daily Bill Amount by Gender')
plt.xlabel('Total Bill')
plt.ylabel('Day')

plt.legend(title='sex', loc='upper right', bbox_to_anchor=(1, 1))

plt.show()
```

**Q08. Create a pie chart based on the data below to show class modality and its weight. Use the output image as reference.** <br>
modality = ['Face to Face', 'Online Anytime', 'Online with Schedule'] <br>
weight = [30, 50, 20]<br><br>
**Output image:**<br>


```python
modality = ['Face to Face', 'Online Anytime', 'Online with Schedule']
weight = [30, 50, 20]
colors = ['yellow', 'orange', 'cyan']
explode = (0.1, 0, 0) # This "pops out" the first slice

plt.pie(weight, labels=modality, colors=colors, explode=explode, 
        autopct='%1.1f%%', shadow=True, startangle=140)

plt.title('Class Modality for Fall 2024')
plt.show()
```

**Q09. Save the the pie chart from Q7 to a pdf file named pie.pdf and save it next to this assignment file**

```python
modality = ['Face to Face', 'Online Anytime', 'Online with Schedule']
weight = [30, 50, 20]
colors = ['yellow', 'orange', 'cyan']
explode = (0.1, 0, 0)

plt.pie(weight, labels=modality, colors=colors, explode=explode, 
        autopct='%1.1f%%', shadow=True, startangle=140)
plt.title('Class Modality for Fall 2024')

plt.savefig('pie.pdf')

plt.show()

print("Success! Your chart has been saved as 'pie.pdf' in your folder.")
```
